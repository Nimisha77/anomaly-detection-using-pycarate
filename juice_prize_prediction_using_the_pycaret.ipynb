```python
#import the dataset from pycaret repository
from pycaret.datasets import get_data
anomaly = get_data('juice')
#import anomaly detection module
from pycaret.anomaly import *
#intialize the setup
exp_ano = setup(anomaly)
```

```python
#Differeny model to use
models()
```

```python
## creating a model
iforest=create_model('iforest')
## plotting a model
plot_model(iforest)
```

```python
## creating a model
knn=create_model('knn')
## plotting a model
plot_model(knn)
```

```python
#use the isolation forest model for predictions
iforest_predictions = predict_model(iforest, data = anomaly)
```
